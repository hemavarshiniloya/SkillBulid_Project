```python
data2=data1.drop('Employee ID',axis=1)
```

```python
data2['Date of Joining']=pd.to_datetime(data['Date of Joining'])
data2['Join Month']=data2['Date of Joining'].apply(lambda x:x.month)
data2['Join Day']=data2['Date of Joining'].apply(lambda x:x.day)
data2=data2.drop('Date of Joining',axis=1)
```

```python
data2['Gender']=data2['Gender'].replace({'Female': 0, 'Male': 1})
data2['Company Type']=data2['Company Type'].replace({'Product': 0, 'Service': 1})
data2['WFH Setup Available'] = data2['WFH Setup Available'].replace({'No': 0, 'Yes': 1})
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
```

```python
X=data2.drop('Burn Rate',axis=1)
y=data2['Burn Rate']
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,shuffle=True)
```

```python
model=RandomForestRegressor()
```

```python
model.fit(X_train,y_train)
```

```python
prediction=model.predict(X_test)
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,prediction)
```
